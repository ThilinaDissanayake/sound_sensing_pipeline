# Reading and visualizing averaged features

```python
features = pd.read_csv("visual_features/"+date+"/vis_features.csv", header=None)
features.index = features[0]
features.drop(0, axis = 1, inplace=True)
```

```python
features
```

```python
def tsne(feat):
    selected_feat = pd.concat([feat.loc[feat.index=="bath_mfcc"],feat.loc[feat.index=="kit_mfcc"], feat.loc[feat.index=="bed_mfcc"]])
    selected_feat.dropna(axis=1, inplace=True)
    display(selected_feat)
    
    
    tsne = TSNE(n_components=2, random_state=150)
    X_2d = tsne.fit_transform(selected_feat.values)

    locations = list(set(selected_feat.index.values))

    loc_list = selected_feat.index.values

    loc_ids = range(len(locations))

    plt.figure(figsize=(7, 7))
    colors = 'r', 'b', 'g', 'c', 'm', 'y'
    markers = ["s", "o", "v", "+", "*", "d"]
    for i, c, label, m in zip(locations, colors, locations, range(len(markers))):
        plt.scatter(X_2d[loc_list == i, 0], X_2d[loc_list == i, 1], c=c, label=label, marker=markers[m])
    plt.legend()
    plt.show()
```
